## Step 4: Organize SNOTEL Data for Bateman, Chamita, and Taos sites

```python
# Build dictionary for the 3 Sites 
# Path to a folder to save the sites and datasets (organize data)


# I am naming my download directory folder "sites":
download_directory = "/Users/oriyavillarreal/Desktop/School Stuff/2022-2023/CEWA 568 Snow Hydrology/snow-tutorial/sites"

# Code below creates the "data" folder if you haven't already
if not os.path.exists(download_directory):
    os.makedirs(download_directory)


sites = {
    # BATEMAN, NM NEAR CARSON NATIONAL FOREST
    "bateman": {
        "local_file": os.path.join(download_directory, "project_bateman.csv"),
        "site_id": "316",
        "dataset": "snotel",
    },
    # CHAMITA, NM NEAR NM/CO BORDER
    "chamita": {
        "local_file": os.path.join(download_directory, "project_chamita.csv"),
        "site_id": "394",
        "dataset": "snotel",
    },
    # TAOS POWDERHORN, NM NEAR TAOS
    "taos": {
        "local_file": os.path.join(download_directory, "project_taos.csv"),
        "site_id": "1168",
        "dataset": "snotel",
    },
}

datasets = {
    "snotel": {
        "data_bateman": os.path.join(download_directory, "project_bateman.csv"),
        "data_chamita": os.path.join(download_directory, "project_chamita.csv"),
        "data_taos": os.path.join(download_directory, "project_taos.csv"),
    },
}
```

```python
# Load site datasets into pandas.
datasets = {
    site: pd.read_csv(sites[site]["local_file"])
    for site in sites.keys()
    for dataset in datasets.keys()
}

#datasets
```

**Note:** The code above allows us to create "key loops" with our data. Since we are working with 3 Sites and selecting data from 3 Periods for each site, the code below allows us to pick and choose the site and data we want to work with. This is really convenient because it saves time when we are organizing our data. Instead of indexing 3 periods for each site for a total of 9 times, the code below let's you only need to do it once for all sites.

### Organize the periods we are working with below:

```python
# start and end dates for 3 periods (1 month Periods)
early_period = pd.date_range("12/01/2022", "12/30/2022")

mid_period = pd.date_range("01/15/2023", "02/15/2023")

late_period = pd.date_range("03/15/2023", "04/15/2023")


def extract_period(dataset: pd.DataFrame, period: pd.DatetimeIndex) -> pd.DataFrame:
    "Extract subset of data based on period start and end."
    dataset.columns = dataset.columns.str.replace(" ", "")
    dataset.date = pd.to_datetime(dataset.date, format="%m/%d/%Y")

    return dataset[dataset.date.isin(period)]
```

Information on code above:

1. We are defining a function called `extract_period` that takes two arguments: 'dataset' and "period." The function operates assuming all our data is compatible with Pandas DataFrame and DatetimeIndex
2. `dataset.columns = dataset.columns.str.replace("","")`: This code line deletes any spaces in the headings of our CSV files
3. `dataset.date = pd.to_datetime(dataset.date, format=%m%d%Y)`: This code line converts the date column for our data to a datetime format. The data is in M/D/Y format.
4. `return dataset[dataset.date.isin(period)]` : This code line returns data based on the specified "period" we are choosing.

## Step 5: Plot our data using a customized Plotting Method

To make plotting multiple time periods for multiple datasets easier and time-efficient, we will explain some tips on how to organize your data and **build plots** based on the code in Step 4 above. The following steps explain the cell below for Step 5:

1. `def build_plot(title: str, dataset: pd.DataFrame) -> None: """ """`: This code defines a function called `build_plot`, which focuses on the title of each plot, and the specific data for each plot.
2. Below the build_plot function code, we begin to plot our data the way we normally do. However, we will plot 3 subplots of the same **TYPE** of data, and we will only write this once.
3. After we write the code for our plots, we will begin to **build** the plots by specifying the dataset (Taos, Bateman, or Chamita) and the time period (Early, Mid Late). #4 is below the cell below.

```python
# Code below only needs to be applied once
def build_plot(title: str, dataset: pd.DataFrame) -> None:
    """ """
    
    
    palette = sns.color_palette("rocket")
    sns.set_palette(palette)
    
    # Now, let's plot our data for our 3 Sites and 3 Periods at each site
    fig, ax = plt.subplots(3, 1, figsize=(15, 15), sharex=True)
    fig.suptitle(title)

    # Plot Snow Depth vs. Air Temperature Data
    ax[0].plot(dataset["airtemp_max"], label="Max Temperature (F)")
    ax[0].plot(dataset["airtemp_min"], label="Min Temperature (F)")
    ax[0].plot(dataset["airtemp_mean"], label="Mean Temperature (F)")
    ax[0].plot(dataset["airtemp_observed"], label="Observed Temperature (F)") 
    ax[0].set_ylabel("Atmospheric Temperature (F)")
    ax[0].tick_params(axis="y")
    ax1 = ax[0].twinx()
    ax1.plot(dataset["snow_depth"], label="Snow Depth (in.)", linestyle='--')
    ax1.set_ylabel("Snow Depth (in.)")
    lines, labels = ax[0].get_legend_handles_labels()
    lines1, labels1 = ax1.get_legend_handles_labels()
    ax[0].legend(lines + lines1, labels + labels1, loc='upper right', bbox_to_anchor=(1.5,1.1))
    ax[0].set_xlabel("Time [Days]")
    fig.tight_layout()

    # Plot Temperature, SWE, and Snow Depth
    ax[1].plot(dataset["snow_depth"], label="Snow Depth (in.)", linestyle='--')
    ax[1].plot(dataset["SWE"], label="SWE (in.)", linestyle='--', marker='p')
    ax[1].set_ylabel("Snow Depth and SWE (in.)")
    ax[1].tick_params(axis="y")
    ax1 = ax[1].twinx()
    ax1.plot(dataset["airtemp_mean"], label="Mean Temperature (F)")
    ax1.set_ylabel("Atmospheric Temperature")
    lines, labels = ax[1].get_legend_handles_labels()
    lines1, labels1 = ax1.get_legend_handles_labels()
    ax[1].legend(lines + lines1, labels + labels1, loc='upper right', bbox_to_anchor=(1.4,1.0))
    ax[1].set_xlabel("Time [Days]")
    fig.tight_layout()

    # Plot Precipitation and SWE
    ax[2].plot(dataset["totalprecip"], label="Precipitation (in)")
    ax[2].set_ylabel("Precipitation Accumulation (in.)")
    ax[2].tick_params(axis="y")
    ax1 = ax[2].twinx()
    ax1.plot(dataset["SWE"], label="SWE (in.)", linestyle='--', marker='p')
    ax1.set_ylabel("SWE (in.)")
    lines, labels = ax[2].get_legend_handles_labels()
    lines1, labels1 = ax1.get_legend_handles_labels()
    ax[2].legend(lines + lines1, labels + labels1, loc='upper right', bbox_to_anchor=(1.4,1.0))
    ax[2].set_xlabel("Time [Days]")
    fig.tight_layout()
```

As explained above, we created a function that allows us to plot our data for any site and any period. We then created our subplots of data we will visualize for each site and period. The **type** of data we are plotting stays the same for each site and period. However, the **datasets** and **time period** is changing based on what we include in the function we built in the first line.

4. `build_plot(title: str, dataset: pd.DataFrame)`: If we wanted to plot 3 subplots of datatypes that we specified above for Taos during the early period, the code would look like this: 
`build_plot("Taos Early Period \n (Dec 1 - Dec 30)", extract_period(datasets["taos"], early_period)) `

* To change the dataset, we will switch out `datasets["taos"]` for `datasets["bateman"]` or `datasets["chamita"]`.
* To change the time period that we want, we will switch out `early_period))` for `mid_period))` or `late_period))`.
